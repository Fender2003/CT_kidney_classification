# resnet

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

load_path = "/Users/dhruv/Dhruv/R_3/ResNet_confusion_matrix.pkl"  

with open(load_path, 'rb') as file:
    loaded_cm = pickle.load(file)

plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)  

sns.heatmap(loaded_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            annot_kws={"size": 16, "weight": "bold"})  

plt.xlabel('Predicted Class', fontsize=18)
plt.ylabel('Actual Class', fontsize=18)

plt.savefig('/Users/dhruv/Dhruv/R_3/confusion_matrix_resnet.pdf', dpi=600, bbox_inches='tight', pad_inches=0)

plt.show()
```

```python
import numpy as np

confusion_matrix = loaded_cm

tp = np.diag(confusion_matrix)
fp = np.sum(confusion_matrix, axis=0) - tp
fn = np.sum(confusion_matrix, axis=1) - tp

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)

macro_average_precision = np.mean(precision)
macro_average_recall = np.mean(recall)
macro_average_f1_score = np.mean(f1_score)

print("Macro-Average Precision:", macro_average_precision)
print("Macro-Average Recall:", macro_average_recall)
print("Macro-Average F1 Score:", macro_average_f1_score)
```

```python
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)
print("Class-wise Precision:", precision)
print("Class-wise Recall:", recall)
print("Class-wise F1 Score:", f1_score)
```

# VGG

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

load_path = "/Users/dhruv/Dhruv/R_3/confusion_matrix_VGG.pkl"  

with open(load_path, 'rb') as file:
    loaded_cm = pickle.load(file)

plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)  

sns.heatmap(loaded_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            annot_kws={"size": 16, "weight": "bold"})  
plt.xlabel('Predicted Class', fontsize=18)
plt.ylabel('Actual Class', fontsize=18)

plt.savefig('/Users/dhruv/Dhruv/R_3/confusion_matrix_VGG.pdf', dpi=600, bbox_inches='tight', pad_inches=0)

plt.show()
```

```python
import numpy as np

# Assuming loaded_cm is your confusion matrix
confusion_matrix = loaded_cm

# Calculate True Positives, False Positives, and False Negatives for each class
tp = np.diag(confusion_matrix)
fp = np.sum(confusion_matrix, axis=0) - tp
fn = np.sum(confusion_matrix, axis=1) - tp

# Calculate Precision, Recall, and F1 Score for each class
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)

# Calculate the macro-average precision, recall, and F1 score
macro_average_precision = np.mean(precision)
macro_average_recall = np.mean(recall)
macro_average_f1_score = np.mean(f1_score)

print("Macro-Average Precision:", macro_average_precision)
print("Macro-Average Recall:", macro_average_recall)
print("Macro-Average F1 Score:", macro_average_f1_score)
```

```python
# Calculate Precision, Recall, and F1 Score for each class
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)
print("Class-wise Precision:", precision)
print("Class-wise Recall:", recall)
print("Class-wise F1 Score:", f1_score)
```
